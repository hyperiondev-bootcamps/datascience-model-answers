```python
# Import pandas
import pandas as pd
```

```python
# Create a DataFrame with balance.txt
df = pd.read_csv("balance.txt", sep=" ")
```

```python
# Return the first five rows
df.head()
```

Write the code needed to produce a report that provides the following information:
* Compare the average income based on ethnicity.
* On average, do married or single people have a higher balance?
* What is the highest income in our dataset?
* What is the lowest income in our dataset?
* How many cards do we have recorded in our dataset? (Hint: use `sum()`)
* How many females do we have information for vs how many males? (Hint: use `count()` for a list of all methods for computation of descriptive stats, explore the [pandas documentation](https://pandas.pydata.org/pandas-docs/stable/reference/frame.html#computations-descriptive-stats)).




### Answers:

#### 1. Compare the average income based on ethnicity.

```python
# Using groupby() to check the average income per ethnicity group
income_per_ethnicity = df.groupby("Ethnicity")["Income"].mean()
print(income_per_ethnicity)
```

```python
# Isolating the average income value of each ethnic group
print("The average income of African American is", df[df["Ethnicity"] == "African American"].loc[:,"Income"].mean())
print("The average income of Asian is", df[df["Ethnicity"] == "Asian"].loc[:,"Income"].mean())
print("The average income of Caucasian is", df[df["Ethnicity"] == "Caucasian"].loc[:,"Income"].mean())
```

#### 2. On average, do married or single people have a higher balance?

```python
balance_per_marital_status = df.groupby("Married")["Balance"].mean()
print(balance_per_marital_status)
```

```python
print("Answer: Overall, single people have a higher balance of " + str(df[df['Married'] == 'No']['Balance'].mean()))
```

#### 3. What is the highest income in our dataset?

```python
print("The highest income in our dataset is", df["Income"].max())
```

#### 4. What is the lowest income in our dataset?

```python
print("The lowest income in our dataset is", df["Income"].min())
```

#### 5. How many cards do we have recorded in our dataset? (Hint: use sum())

```python
sum_cards = df["Cards"].sum()
print(f"We have recorded in our dataset {sum_cards} cards.")
```

#### 6. How many females do we have information for vs how many males? (Hint: use count() for a list of all methods for computation of descriptive stats, explore the pandas documentation).

```python
# Counting the number of females and males in our dataset using .vale_counts()
df["Gender"].value_counts()
```

```python
# Isolating the counting of females and males
print("We have information of", df[df["Gender"] == "Female"].loc[:,"Gender"].count(), "Female")
print("We have information of", df[df["Gender"] == "Male"].loc[:,"Gender"].count(), "Male")
```
