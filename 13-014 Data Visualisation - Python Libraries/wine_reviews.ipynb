```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load data
# By setting index_col=[0], you use the first column of the CSV file as the index of the DataFrame,
# avoiding creation of the 'Unnamed: 0' column.
wine = pd.read_csv("wine.csv", index_col=[0])
wine.head()
```

```python
# Check the basic information of the DataFrame
wine.info()
```

```python
# Check the unique values in the 'variety' column
wine["variety"].unique()
```

```python
# Filter the dataset to only contain "Cabernet Sauvignon", "Pinot Noir", and "Chardonnay" wines
filtered_data = wine[wine["variety"].isin(["Cabernet Sauvignon", "Pinot Noir", "Chardonnay"])]
filtered_data
```

```python
# Check the count of each wine variety in the filtered DataFrame
filtered_data["variety"].value_counts()
```

```python
# Create a multi-plot grid using seaborn FacetGrid
# Each grid will represent one of the three wine varieties: "Cabernet Sauvignon", "Pinot Noir", and "Chardonnay"
# In each grid, show a histogram of the "points" column
graph = sns.FacetGrid(filtered_data, col="variety")
graph.map(sns.histplot, "points")
plt.show()
```
